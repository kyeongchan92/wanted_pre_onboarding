입력받기

```python
import numpy as np

# 문자열 집합 S가 N(코드에서는 n)개 주어짐 : "aaa","aaaabb", "aabbcc", "abb", "bcc", "bbcc", "aabbccdd", "aaabb", "cccd"
# s_strings에 저장
n = int(input())
s_strings = []
for _ in range(n):
    s_strings.append(input())

# 대상 문자열이 q개 주어짐, q_strings에 저장
q = int(input())  
q_strings = []
for _ in range(q):  # "aaaabbaabbccdd"
    q_strings.append(input())
```

# 1번 문제

```python
for q in q_strings:
    print(', '.join(['YES' if s in q  else 'NO' for s in s_strings]))
```

# 2번 문제

```python
for q in q_strings:
    print(f"대상 문자열 : {q if len(q) < 20 else q[:20]+'.....'}")
    for s in s_strings:
        split_idx = 0
        if s in q:
            pos = q.find(s)  # string.find()는 가장 처음 매칭되는 position만 반환
            print(f'#pos = {pos}, pattern = {s}')
    print('-' * 50)
```

# 3번 문제

```python
for q in q_strings:
    print(f"대상 문자열 : {q if len(q) < 20 else q[:20]+'.....'}")
    for s in s_strings:
        split_idx = 0
        target = q[:]
        while s in target:  # 대상 문자열을 줄여나감. target = 확인한 position은 제외한 그 뒤부터의 문자열
            pos = target.find(s)
            print(f'#pos = {split_idx + pos}, pattern = {s}')
            split_idx += pos + 1
            target = q[split_idx:]
    print('-' * 50)
```
